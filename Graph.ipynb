```python
# importing library
import matplotlib.pyplot as plt

# function to add value labels
def addlabels(x,y):
	for i in range(len(x)):
		plt.text(i, y[i]//2,y[i], ha = 'center',
				Bbox = dict(facecolor = 'yellow', alpha = 1.5))

if __name__ == '__main__':
	
	# creating data on which bar chart will be plot
	x = [ "Dhubri Quake(30)", "Bihar-Nepal Quake(34)", "Assam Quake(50)", "Chittagong Earthquake(97)",
        "Bhuj Earthquake(01)", "Chittagong Rangamati(03)", "Bangladesh Myanmar(13)"]
	y = [7.1, 8.3, 8.5, 6.1, 7.9, 5.9, 5.9]
	
	# setting figure size by using figure() function
	plt.figure(figsize = (20,5))
	
	# making the bar chart on the data with color red
	plt.bar(x, y, color = 'red')
	
	# calling the function to add value labels
	addlabels(x, y)
	
	# giving title to the plot
	plt.title("Last 100 years Earthquack")
	
	# giving X and Y labels
	plt.xlabel("Incident Name")
	plt.ylabel("Richter")
	
	# visualizing the plot
	plt.show()
```

```python
# importing library
import matplotlib.pyplot as plt

# function to add value labels
def addlabels(x,y):
	for i in range(len(x)):
		plt.text(i, y[i]//2,y[i], ha = 'center',
				Bbox = dict(facecolor = 'yellow', alpha = 1.5))

if __name__ == '__main__':
	
	# creating data on which bar chart will be plot
	x = ["Chittagong(07)", "Chittagong(08)", "Coxs Bazar(10)",
         "Chittagong,Coxs Bazar,Bandarban(12)", "Lalkhan,Chittagong(13)",]
	y = [127, 11, 96, 122, 2]
	
	# setting figure size by using figure() function
	plt.figure(figsize = (20,7))
	
	# making the bar chart on the data with color red
	plt.bar(x, y, color = 'blue')
	
	# calling the function to add value labels
	addlabels(x, y)
	
	# giving title to the plot
	plt.title("Recent major landslides in Bangladesh")
	
	# giving X and Y labels
	plt.xlabel("Place & Year")
	plt.ylabel("Death")
	
	# visualizing the plot
	plt.show()
```
